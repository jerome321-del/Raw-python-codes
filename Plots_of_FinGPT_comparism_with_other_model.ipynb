```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

datasets = [
    "flare-FPB", "FiQA-SA", "Headline", "FIN", "ConvFinQA",
    "FinQA", "CIKM18", "StockNet", "BigData22", "ECTSum"
]

fingpt = np.array([87.62, 95.80, 95.50, 69.76, 28.47, 3.80, 45.00, 48.22, 52.74, 0.00])
finma  = np.array([87.00, 79.00, 97.00, 69.00, 20.00, 4.00, 53.00, 56.00, 49.00, 8.00])

diff = fingpt - finma

datasets_plot = datasets[::-1]
diff_plot = diff[::-1]
colors = ["orange" if x > 0 else "blue" for x in diff_plot]

# White background
fig, ax = plt.subplots(figsize=(12, 8), facecolor="white")
ax.set_facecolor("white")

y = np.arange(len(datasets_plot))
bars = ax.barh(y, diff_plot, color=colors, height=0.8)

# Zero line and grid
ax.axvline(0, color="black", linewidth=1.5)
ax.xaxis.grid(True, linestyle="--", color="gray", alpha=0.8)

# Bar labels
for i, v in enumerate(diff_plot):
    label = f"{abs(v):.2f}%"
    if v > 0:
        ax.text(v / 2, i, label, va="center", ha="center", color="white", fontsize=12)
    elif v < 0:
        ax.text(v / 2, i, label, va="center", ha="center", color="white", fontsize=12)
    else:
        ax.text(0.15, i, "0.00%", va="center", ha="left", color="black", fontsize=10)

# Axis labels in black
ax.set_yticks(y)
ax.set_yticklabels(datasets_plot, color="black", fontsize=11)
ax.set_xlabel("Performance Difference (FinGPT - FinMA 7B) [%]", color="black", fontsize=12)
ax.set_title("FinGPT vs FinMA 7B Across Financial NLP Tasks", color="black", fontsize=15)

ax.tick_params(axis="x", colors="black")
ax.tick_params(axis="y", colors="black")

# Optional: make axis border black
for spine in ax.spines.values():
    spine.set_color("black")

legend_elements = [
    Patch(facecolor="orange", label="FinGPT > FinMA 7B"),
    Patch(facecolor="blue", label="FinMA 7B > FinGPT")
]
leg = ax.legend(handles=legend_elements, loc="lower right", frameon=True, facecolor="white", edgecolor="black")
for text in leg.get_texts():
    text.set_color("black")

plt.tight_layout()
plt.savefig("fingpt_vs_finma_comparison.png", dpi=300, facecolor="white")
plt.show()
```
